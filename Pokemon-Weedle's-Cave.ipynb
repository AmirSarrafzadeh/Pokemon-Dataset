# Ploting with Matplotlib

- line: 
    when axis is time
- Scatter:
    When there is correlation between two variable
- Histogram:
    when need to see the numerical data distribution

### Line Plot

```python
plt.figure(figsize=(14,6))
pokemon.Speed.plot(kind='line', color='g', label='Speed', linewidth=1, alpha=1, grid=True, linestyle=':')
pokemon.Defense.plot(kind='line', color='r', label='Defence', linewidth=1, alpha=1, grid=True, linestyle=':')

plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot example")
#plt.show()
```

### Scatter Plot

```python
# x = Attack, y= Defense
plt.figure(figsize=(10,8))
pokemon.plot(kind='scatter', x='Attack', y='Defense', alpha=.5, color='b')

plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack - Defense Scatter Fig:')
plt.show()
```

### Histogram

```python
# bins = # of bars in the Fig

pokemon.Speed.plot(kind='hist', bins=50,figsize=(14,6), color='silver')
```

```python
# clf : cleans the graph so that it did not show
pokemon.Defense.plot(kind='hist',bins=10)
plt.clf()
```
